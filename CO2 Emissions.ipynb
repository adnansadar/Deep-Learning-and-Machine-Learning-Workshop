```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
test_df = pd.read_csv('co2.txt')
```

```python
test_df
```

```python
df_new=pd.DataFrame(test_df)
```

```python
plt.bar(df_new['Country'],df_new['CO2 emissions (kt) in 2014[2]']/10)
plt.xticks(rotation='90')
plt.yticks(rotation='45')
plt.show()
```

```python
slices=df_new['CO2 emissions (kt) in 2014[2]']
activities=df_new['Country']
plt.figure(figsize=(5,5))
plt.pie(slices[:10],
       labels=activities[:10],
       wedgeprops={'linewidth':3},
       radius=2,
       startangle=90,
       shadow='True')
plt.title("CO2 Emissions",fontweight='bold')
plt.legend(loc=3)
plt.show()
```

```python
x=np.arange(1,5)
y=x**2-5
y1=x**2-7
plt.title("Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.plot(x,y,label='f1')
plt.plot(x,y1,label='f2')
plt.legend(loc=7) #sets position of legend on the graph
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,label='f1')
plt.legend()
plt.subplot(1,2,2)
plt.plot(x,y1,label='f2',color='yellow')
plt.legend()
plt.show()
```
